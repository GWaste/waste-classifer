```python
#@title Pengaturan

# hardcode, gak perlu diubah
savedModelPath = "drive/Shareddrives/B21-CAP0331/models/"
tfLiteFolderPath = "drive/Shareddrives/B21-CAP0331/models_tflite/"

#@markdown Nama model
modelName = "2021-05-21_14-52_model-build-v3_resnet50v2_fine-tune_50epochs" #@param {type:"string"}
```

```python
import tensorflow as tf

# Convert the model
saved_model_dir = savedModelPath + modelName
converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_dir) # path to the SavedModel directory
tflite_model = converter.convert()

# Save the model.
tflitePath = tfLiteFolderPath + modelName + ".tflite"
with open(tflitePath, 'wb') as f:
  f.write(tflite_model)

print("Saved to", tflitePath)
drive.flush_and_unmount()
drive.mount("drive")
```
